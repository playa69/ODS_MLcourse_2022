```python
# v1 (long)
j = 0 #counter-controlled repitition
with open("output_v2.txt", "w") as new_file:
    with open("import_file_2_1.txt", "r") as file_1:
        with open("import_file_2_2.txt", "r") as file_2:
            array = []
            
            for line2 in file_2:
                lst = line2.split()
                array.append(lst[0])
                array.append(lst[1])
            
            i = 0
            
            for line1 in file_1:
                x = array[i]
                y = int(array[i+1]) + 1                
                #print( line1[int(x):int(y)] )                
                i = i+2
                new_file.write(line1[int(x):int(y)])
                j+=1
                if j != 3000:
                    new_file.write(' ')
                   
```

```python
# v2 (faster)

LINES = 3000

with open("output.txt", "w+") as new_file:
    with open("import_file_2_1.txt", "r") as file_1:
        with open("import_file_2_2.txt", "r") as file_2:
            for i in range(LINES):
                s = file_1.readline()
                print(s)
                start, end = map(int, file_2.readline().split())
                # [int(x) for x in s.split()]            
                # \n in the end doesn't hurt
                new_file.write(s[start:end + 1])           
                if i != LINES - 1:
                    new_file.write(' ')
new_file.close()
file_1.close()
file_2.close()
                    
                    
                    
```
